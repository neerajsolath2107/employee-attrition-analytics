```python
total_emp = df.shape[0]
attrition_count = df['Attrition'].value_counts()
resigned_emp = attrition_count['Yes']
staying_emp = attrition_count['No']
attrition_rate = (resigned_emp / total_emp ) * 100

print("Total Employees:", total_emp)
print("Staying Employees:", staying_emp)
print("Resigned Employees:", resigned_emp)
print("Attrition Rate (%):", round(attrition_rate, 2))
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['Attrition_numeric'] = df['Attrition'].map({'Yes' : 1, 'No' : 0})
```

```python
# Age Groups Create करना
bins = [18, 25, 35, 45, 60]
labels = ['18-25', '26-35', '36-45', '45+']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# AgeGroup wise resignation देखना
age_attrition = df.groupby('AgeGroup')['Attrition_numeric'].mean() * 100
print("\nAttrition by Age Group (%):\n", age_attrition)
```

```python
# Salary Groups बनाना
income_bins = [0, 3000, 7000, 20000]
income_labels = ['Low', 'Medium', 'High']

df['IncomeGroup'] = pd.cut(df['MonthlyIncome'], bins=income_bins, labels=income_labels)

income_attrition = df.groupby('IncomeGroup')['Attrition_numeric'].mean() * 100
print("\nAttrition by Income Group (%):\n", income_attrition)
```

```python
dept_attrition = df.groupby('Department')['Attrition_numeric'].mean() * 100
print("\nAttrition by Department (%):\n", dept_attrition)
overtime_attrition = df.groupby('OverTime')['Attrition_numeric'].mean() * 100
print("\nOvertime Impact on Attrition (%):\n", overtime_attrition)

df.to_csv("cleaned_attrition_data.csv",index = False)
```

```python
import matplotlib.pyplot as plt

attrition_pie = df['Attrition'].value_counts()

plt.figure()
plt.pie(attrition_pie, labels=attrition_pie.index, autopct='%1.1f%%')
plt.title("Overall Attrition Distribution")
plt.show()
```

```python
plt.figure()
df.groupby('AgeGroup')['Attrition_numeric'].mean().plot(kind='bar')
plt.title("Attrition by Age Group")
plt.ylabel("Attrition Rate")
plt.show()
```

```python
plt.figure()
df.groupby('Department')['Attrition_numeric'].mean().plot(kind='bar')
plt.title("Department Wise Attrition")
plt.ylabel("Attrition Rate")
plt.show()
```

```python
plt.figure()
plt.hist(df['MonthlyIncome'])
plt.title("Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.show()
```
