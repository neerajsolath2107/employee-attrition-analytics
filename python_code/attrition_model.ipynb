```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("/kaggle/input/employee-attrition/WA_Fn-UseC_-HR-Employee-Attrition.csv.xls")

# Attrition को numeric में convert करें
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Target Variable
y = df['Attrition']

# Feature Columns (Attrition छोड़कर बाकी)
X = df.drop(columns=['Attrition'])
```

```python
# सभी categorical columns को label encode करें
encoder = LabelEncoder()

for column in X.columns:
    if X[column].dtype == 'object':
        X[column] = encoder.fit_transform(X[column])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Training Data Size:", X_train.shape[0])
print("Testing Data Size:", X_test.shape[0])
```

```python
log_model = LogisticRegression()

# Model Train करें
log_model.fit(X_train, y_train)

# Prediction
y_pred_log = log_model.predict(X_test)

# Accuracy Check
print("\nLogistic Regression Accuracy:", accuracy_score(y_test, y_pred_log))
```

```python
rf_model = RandomForestClassifier()

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("\nRandom Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
print("\nConfusion Matrix (Random Forest):\n", confusion_matrix(y_test, y_pred_rf))
```

```python
print("\nDetailed Report:\n", classification_report(y_test, y_pred_rf))
```

```python
# Target और Features अलग करना
y = df['Attrition']

X = df.drop(columns=['Attrition'])

# Categorical columns encode करना
encoder = LabelEncoder()

for column in X.columns:
    if X[column].dtype == 'object':
        X[column] = encoder.fit_transform(X[column])

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
import pandas as pd

# Probability निकालना
probabilities = rf_model.predict_proba(X_scaled)

# Prediction निकालना
predicted = rf_model.predict(X_scaled)

# Integrated Result DataFrame
prediction_df = pd.DataFrame({
    "EmployeeNumber": df["EmployeeNumber"],
    "Predicted_Attrition": predicted,
    "Attrition_Risk_Probability": probabilities[:, 1]
})

# CSV Export
prediction_df.to_csv("attrition_predictions.csv", index=False)

print("attrition_predictions.csv successfully created")
```
